```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
scale_factor = 50;
noise_level = 350;
```

```python
# Generating random data
x = np.random.randint(0,100,size=50);

# Generating y such that its slightly proportional to x
scaled_array = x * scale_factor    
noise = np.random.randn(len(x)) * noise_level
y = scaled_array + np.random.rand(50)*noise
```

```python
print(x)
print(y)
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(x, y,marker='o',color='b')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Linear Graph')

plt.legend()

plt.grid(True)
plt.show()
```

```python
def compute_MSE(x,y,w,b):
    cost = 0
    m = x.shape[0]

    for i in range(m):
        y_hat = w*x[i] + b
        cost = cost + (y[i]-y_hat)**2
    
    total_cost = (1/(2*m))*cost
    return total_cost 
```

```python
def compute_gradient(x,y,w,b):
    partial_dw = 0
    partial_db = 0
    m = x.shape[0]

    for i in range(m):
        y_hat = w*x[i] + b
        curr_dw = (y_hat - y[i])*x[i]
        curr_db = (y_hat - y[i])
        partial_dw+=curr_dw
        partial_db+=curr_db

    partial_db /= m
    partial_dw /= m
    return partial_dw,partial_db
```

```python
def gradient_descent(x,y,w,b,alpha,num_iter):
    for i in range(num_iter):
        partial_dw,partial_db = compute_gradient(x,y,w,b)
        w = w - alpha*partial_dw
        b = b - alpha*partial_db

    return w,b
```

```python
x = x.astype(np.float64)
y = y.astype(np.float64)
# x = (x - np.mean(x)) / np.std(x)  # Standardization
# y = (y - np.mean(y)) / np.std(y)  # Standardization
print(x)
print(y)
```

```python
w = 0
b = 0
num_iter = 10000
alpha = 0.01

w,b = gradient_descent(x,y,w,b,alpha,num_iter)
print(f"w and b fonud after gradient descent are {w} , {b}")
```

```python
compute_MSE(x,y,w,b)
```

```python
x_sample = np.linspace(-1.5,2,50)
y_sample = w*x_sample + b

plt.figure(figsize=(8, 5))
plt.scatter(x, y,marker='o',color='b')
plt.plot(x_sample,y_sample)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Linear Graph')

plt.legend()

plt.grid(True)
plt.show()
```
